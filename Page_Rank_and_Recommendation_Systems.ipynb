### 1.1.4
Create the adjacency matrix for the graph, assigning to each character a node identifier equal to the index that the character name has in ascending alphabetical order (remember that the first element of a list in Python has index 0).

```python
#YOUR CODE STARTS HERE#
# dictiory that map id to character name
map_character_id_character_name = {i: x for i,x in enumerate(character_names_list)}
# reverse of the previous dictionary
map_character_name_character_id = {x: i for i,x in enumerate(character_names_list)}

# construct the Adjancy matrix
adj = pd.crosstab(Harry_Potter_graph.character_1, Harry_Potter_graph.character_2)
idx = character_names_list
adj = adj.reindex(index = idx, columns=idx, fill_value=0)
adj = adj + np.transpose(adj)

adj_of_Harry_Potter_graph = adj.to_numpy()





#YOUR CODE ENDS HERE#
#THIS IS LINE 20#
```

### 1.1.5
Compute the PageRank vector for the obtained graph using a damping factor of 0.8.

```python
#YOUR CODE STARTS HERE#
# Set the parameters for the PageRank computation
alpha = 0.2
damping_factor = 1. - alpha

# PageRank
pagerank = PageRank(damping_factor=damping_factor, solver="piteration", 
                    n_iter=1000, tol=10**-6)

# Compute Page-Rank on the entire Harry Potter graph
pagerank_vector = pagerank.fit_transform(adj_of_Harry_Potter_graph)







#YOUR CODE ENDS HERE#
#THIS IS LINE 20#
```

Print the name and the PageRank score of the top-10 characters according to the PageRank score of their associated nodes.

```python
#YOUR CODE STARTS HERE#
k = 10
ind = np.argpartition(pagerank_vector, -k)[-k:]
ind  = ind[np.argsort(pagerank_vector[ind])][::-1]
top_10 = [(map_character_id_character_name[x], np.round(pagerank_vector[x],6)) for x in ind]

display(pd.DataFrame(top_10, columns = ['Character name', 'Pagerank score']))

#YOUR CODE ENDS HERE#
#THIS IS LINE 10#
```
